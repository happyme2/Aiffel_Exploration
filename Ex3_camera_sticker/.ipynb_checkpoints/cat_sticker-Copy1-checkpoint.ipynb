# EXP 3 ) Cat_sticker

# 이미지 불러오기

```python
import cv2             #이미지 처리를 위해 opencv 읽기
import matplotlib.pyplot as plt         #이미지 출력을 위한 matplotlib 읽기
import numpy as np

#이미지 읽기
import os
my_image_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/iu.jpg'
img_bgr = cv2.imread(my_image_path)    #- OpenCV로 이미지를 읽기 (opencv는 bgr사용)
img_bgr = cv2.resize(img_bgr, (1000, 950))    # 이미지 크기 Resize
img_show = img_bgr.copy()      #- 출력용 이미지 img_show함수에 넣기
plt.imshow(img_bgr) #bgr이라 이미지 색 변경 (r->b,b->r)
plt.show()
```

# 이미지 검출

```python
import dlib
detector_hog = dlib.get_frontal_face_detector()   #hog detector 선언

img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB) #이미지 rgb로 불러오기
dlib_rects = detector_hog(img_rgb, 1)   #- (image, num of img pyramid)

print(dlib_rects)   # 찾은 얼굴영역 좌표

for dlib_rect in dlib_rects:
    l = dlib_rect.left()
    t = dlib_rect.top()
    r = dlib_rect.right()
    b = dlib_rect.bottom()

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA)

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

# 얼굴 랜드마크

```python
import os
model_path = os.getenv('HOME')+'/aiffel/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)

list_landmarks = []
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)     #RGB 이미지를 입력받고 dlib.full_object_detection 반환
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))     # tuple (x,y) 68개로 이루어진 리스트 
    list_landmarks.append(list_points)    #68개의 랜드마크가 얼굴 개수만큼 저장된다.

print(len(list_landmarks[0]))      #랜드마크 갯수

for landmark in list_landmarks:
    for idx, point in enumerate(list_points):
        cv2.circle(img_show, point, 2, (0, 255, 255), -1) # yellow

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

# 스티커 붙이기

```python
# 얼굴 좌표 확인하기
for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    print (landmark[33]) # nose center index : 30
    x = landmark[33][0]
    y = landmark[33][1] - dlib_rect.width()//2
    w = dlib_rect.width()
    h = dlib_rect.width()
    print ('(x,y) : (%d,%d)'%(x,y))
    print ('(w,h) : (%d,%d)'%(w,h)) 
    
```

```python
#스티커 이미지 불러오기
import os
sticker_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/cat-whiskers.png'
img_sticker = cv2.imread(sticker_path)
img_sticker = cv2.resize(img_sticker, (w,h))
print (img_sticker.shape)       # 코 스티커 w,h 로 resize 

plt.imshow(img_sticker)
plt.show()
```

```python
#이미지 좌표 찾기
refined_x = x - w // 2  # (x_left) 원래 위치보다 w/2 만큼 왼쪽으로 이동
refined_y = y      # (y_top) 
print ('(x,y) : (%d,%d)'%(refined_x, refined_y))
```

```python
#원본 이미지에 스티커 맞추기
sticker_area = img_show[refined_y:refined_y+img_sticker.shape[0], 
                        refined_x:refined_x+img_sticker.shape[1]]
img_show[refined_y:refined_y+img_sticker.shape[0], 
         refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8)    
            # (고양이 수염 이미지==흰색 부분은 얼굴 코 픽셀, 아닌 부분은 고양이 스티커 픽셀) -> 합쳐서 고양이 코를 붙인 얼굴 
    
plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show()
```

### **1) 얼굴 각도가 달라질때**  
  
  - 얼굴 각도가 바뀌더라도, 눈코입이 뚜렷히 나타나게 되면 얼굴이 잘 인식된다.  
  - 하지만 스티커를 붙이는 과정에서 얼굴의 각도에 맞춰 스티커의 각도가 변하지 않는 현상이 나타났다. 검색을 통하여 이미지의 각도를 warpAffine 함수와, getRotationMatrix2D 함수를 통하여 변환한다는 과정은 알게 되었으나, 적용하려는 부분에서 이해가 어려웠다.  

```python
# 이미지 불러오기
import cv2              #이미지 처리를 위해 opencv 읽기
import matplotlib.pyplot as plt         #이미지 출력을 위한 matplotlib 읽기
import numpy as np

#이미지 읽기
import os
my_image_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/iu_side1.jpg'
img_bgr = cv2.imread(my_image_path)    #- OpenCV로 이미지를 읽기 (opencv는 bgr사용)
img_bgr = cv2.resize(img_bgr, (1100, 1200))    # 이미지 크기 Resize
img_show = img_bgr.copy()      #- 출력용 이미지 img_show함수에 넣기
```

```python
#이미지 검출
import dlib
detector_hog = dlib.get_frontal_face_detector()   #hog detector 선언

img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB) #이미지 rgb로 불러오기
dlib_rects = detector_hog(img_rgb, 1)   #- (image, num of img pyramid)

print(dlib_rects)   # 찾은 얼굴영역 좌표

for dlib_rect in dlib_rects:
    l = dlib_rect.left()
    t = dlib_rect.top()
    r = dlib_rect.right()
    b = dlib_rect.bottom()

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA)

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
```

```python
#얼굴 랜드마크 지정
import os
model_path = os.getenv('HOME')+'/aiffel/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)

list_landmarks = []
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)     #RGB 이미지를 입력받고 dlib.full_object_detection 반환
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))     # tuple (x,y) 68개로 이루어진 리스트 
    list_landmarks.append(list_points)    #68개의 랜드마크가 얼굴 개수만큼 저장된다.

print(len(list_landmarks[0]))      #랜드마크 갯수

for landmark in list_landmarks:
    for idx, point in enumerate(list_points):
        cv2.circle(img_show, point, 2, (0, 255, 255), -1) # yellow

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

```python
# 스티커 붙이기
# 얼굴 좌표 확인하기
for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    print (landmark[33]) # nose center index : 30
    x = landmark[33][0]
    y = landmark[33][1] - dlib_rect.width()//2
    w = dlib_rect.width()
    h = dlib_rect.width()
    print ('(x,y) : (%d,%d)'%(x,y))
    print ('(w,h) : (%d,%d)'%(w,h)) 
    
```

```python
#스티커 이미지 불러오기
import os
sticker_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/cat-whiskers.png'
img_sticker = cv2.imread(sticker_path)
img_sticker = cv2.resize(img_sticker, (w,h))
print (img_sticker.shape)       # 코 스티커 w,h 로 resize 

#이미지 좌표 찾기
refined_x = x - w // 2  # (x_left) 원래 위치보다 w/2 만큼 왼쪽으로 이동
refined_y = y      # (y_top) 
print ('(x,y) : (%d,%d)'%(refined_x, refined_y))
```

```python
#원본 이미지에 스티커 맞추기(자연스럽게 적용)
sticker_area = img_bgr[refined_y:refined_y+img_sticker.shape[0], 
                        refined_x:refined_x+img_sticker.shape[1]]
img_bgr[refined_y:refined_y+img_sticker.shape[0], 
         refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8)    
            # (고양이 수염 이미지==흰색 부분은 얼굴 코 픽셀, 아닌 부분은 고양이 스티커 픽셀) -> 합쳐서 고양이 코를 붙인 얼굴 
    
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.show()
```

- **눈코입이 가려진 사진** : 1) 선글라스를 써도 안에 눈코입이 뚜렷히 확인된다면 얼굴이 인식된다.  
                   2)하지만 눈이 보이지 않는다면 얼굴 인식이 되지 않는 것을 알 수 있다.

```python
#이미지 검출 / 이미지 불러오기
my_image_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/sunglasses.jpg'
img_bgr = cv2.imread(my_image_path)    #- OpenCV로 이미지를 읽기 (opencv는 bgr사용)
img_bgr = cv2.resize(img_bgr, (1100, 1000))    # 이미지 크기 Resize
img_show = img_bgr.copy()      #- 출력용 이미지 img_show함수에 넣기
plt.imshow(img_bgr) #bgr이라 이미지 색 변경 (r->b,b->r)
#이미지 검출
detector_hog = dlib.get_frontal_face_detector()   #hog detector 선언

img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB) #이미지 rgb로 불러오기
dlib_rects = detector_hog(img_rgb, 1)   #- (image, num of img pyramid)

print(dlib_rects)   # 찾은 얼굴영역 좌표

for dlib_rect in dlib_rects:
    l = dlib_rect.left()
    t = dlib_rect.top()
    r = dlib_rect.right()
    b = dlib_rect.bottom()

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA)

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)

model_path = os.getenv('HOME')+'/aiffel/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)

list_landmarks = []
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)     #RGB 이미지를 입력받고 dlib.full_object_detection 반환
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))     # tuple (x,y) 68개로 이루어진 리스트 
    list_landmarks.append(list_points)    #68개의 랜드마크가 얼굴 개수만큼 저장된다.

print(len(list_landmarks[0]))      #랜드마크 갯수

for landmark in list_landmarks:
    for idx, point in enumerate(list_points):
        cv2.circle(img_show, point, 2, (0, 255, 255), -1) # yellow

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```
